```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

%config InlineBackend.figure_format = 'retina' 

n, p = 40, 0.4
x = np.arange(0, n + 1)
pmf = binom.pmf(x, n, p)

fig, ax = plt.subplots(figsize=(10, 5))
ax.bar(x, pmf, color='#3498db', alpha=0.8, edgecolor='white')

ax.set_title(f'Binomial PMF (n={n}, p={p})')
ax.set_xlabel('Number of Successes')
ax.set_ylabel('Probability')
plt.show()
```

```python
from ipywidgets import interact, FloatSlider, IntSlider

def plot_binomial(n, p):
    x = np.arange(0, n + 1)
    pmf = binom.pmf(x, n, p)
    
    plt.figure(figsize=(10, 5))
    plt.bar(x, pmf, color='teal', alpha=0.6)
    plt.ylim(0, 0.5) 
    plt.title(f'Binomial Distribution: n={n}, p={p:.2f}')
    plt.show()

interact(plot_binomial, 
         n=IntSlider(min=1, max=100, step=1, value=20), 
         p=FloatSlider(min=0, max=1, step=0.01, value=0.5));
```

```python
from scipy.stats import norm

n, p = 50, 0.5
x = np.arange(0, n + 1)
pmf = binom.pmf(x, n, p)

mu = n * p
sigma = np.sqrt(n * p * (1 - p))
x_norm = np.linspace(0, n, 100)
y_norm = norm.pdf(x_norm, mu, sigma)

plt.bar(x, pmf, alpha=0.5, label='Binomial PMF')
plt.plot(x_norm, y_norm, 'r-', label='Normal Approximation')
plt.legend()
plt.show()
```
